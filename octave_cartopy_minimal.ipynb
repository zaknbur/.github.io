f1n = 'survey_data_file.csv';

n_lines = 0;
latitude=[];

while(~feof(f1))
  str=fgetl(f1);
  C = textscan(str,'%d,%s,%s,%d,%d,%f,%f,%f,%s,%s,%f,%d,','delimiter',',');
  
  if (isnumeric(C{1}) && isnumeric(C{1,6}) && ~isnan(C{1,6}) && ~isnan(C{1,7}))

end

longitude_i = min(longitude):0.0001:max(longitude);

[longitude_m,latitude_m] = meshgrid(longitude_i,latitude_i);

figure
p=[longitude latitude];
t=delaunayn(p);
fi = tinterp(p,t,tmi,longitude_m,latitude_m,'quadratic');
pcolor(longitude_m,latitude_m,fi); hold on
plot(longitude,latitude)
